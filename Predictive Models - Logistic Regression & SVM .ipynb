# Model 1: Logistic Regression Model

## DATA PRE-PROCESSING

## 1. Importiing the libraries

```python
import numpy as np   #library for mathematical operations in python
import matplotlib.pyplot as plt   #Its a plotting library used for 2D graphics
import pandas as pd  #Library to import datasets and manage them
```

## 2. Importing Dataset

```python
dataset = pd.read_excel("C:/Users/NIKITA/Desktop/Nucleus/Nucleus.xlsx")
```

```python
dataset
```

## 3. To Determine Dependent & Independent variables

```python
X = dataset.iloc[:, :-1].values 
y = dataset.iloc[:, 16].values 
```

```python
X
```

```python
y
```

## 4. Treating missing values in dataset

```python
from sklearn.preprocessing import Imputer   
imputer_1 = Imputer(missing_values = 'NaN', strategy = 'mean', axis=0)
imputer_1 = imputer_1.fit(X[:, 1:3])
X[:, 1:3] = imputer_1.transform(X[:, 1:3])
```

## 5. Splitting the Dataset into training set & test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

## 6. Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
X_train
```

```python
X_test
```

## Fitting Logistic Regression to the Training Data

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

## Predicting the test set results

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

## Making confusion matrix- Logistic Regression

```python
from sklearn.metrics import confusion_matrix
confu_m = confusion_matrix(y_test,y_pred)
```

```python
confu_m
```
